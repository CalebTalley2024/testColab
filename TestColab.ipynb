```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create a plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')

# Display the plot
plt.show()
```
